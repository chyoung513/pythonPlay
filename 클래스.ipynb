# 클래스

## 특징
    1. 묶음
        - 변수(멤버변수, 필드, 속성, 인스턴스 변수)
        - 함수(메서드, 멤버 함수)
        
    2. 설계도(추상적)
        - 객체(인스턴스) : 실제 작업 가능한 메모리 공간
        
    3. 사용자 정의 자료형
     

## 작성방법
    class 이름 [...]:
        [멤버 변수]
        [메서드]
        
## 생성자와 소멸자

    1) 메서드 ( = 함수)
    2) 객체를 생성하는 용도
    3) 객체를 처음 만들때 호출되고 단 한번만 호출할 수 있음
    4) 초기화 용도
    5) 소멸자는 객체가 소멸되기 직전에 호출됨
    6) callback
    
## 접근권한(Access Modifier)

    1) public : default
    2) private : 이름 앞에 __ 사용
        - 간접적으로 사용을 할 수 있게 하기 위해서는 getter, setter 메서드를 준비해야 한다.

## 상속
    1) 종류
        - 상속 :하나의 부모로부터 물려받는 구조
        - 다중 상속 : 여러 부모로부터 물려받는 구조
        
    2) 장점
        - 코드의 재활용
            시간, 비용 절감
            견고한 코드 작성(검증받은 코드이기 때문에)
            코드 간결
            
    3) 설계시 주의할 점
        부모 클래스는 모든 자식이 공통으로 가질 수 있는 기능만을 최소한의 설계로 작성
        자식 클래스는 그 외에 필요한 기능은 추가
    
    4) UML
        하얀 바탕의 삼각 화살표
     
     5) Override(재정의)
         - 부모로 물려받은 메서드를 입맛에 맞게 수정하는 작업
         - 다형성
         
     6) 추상클래스
         - 추상 메서드를 1개이상 보유
         - 추상 메서드란?
             내용이 없이 정의만 되어있는 메서드
             상속 받는 자식 클래스에서는 반드시 구현해야 하는 강제성을 가지고 있다.
         - 추상 클래스는 객체를 생성할 수 없다.
             상속을 위한 용도로만 사용
             명확하게 용도를 지정함으로써 필요없는 객체가 생성되는 것을 방지
         - 해야 할 작업에 대해 실수를 방지하고 명확한 의도를 갖고 구현할 수 있도록 유도
        
      7) object : 파이썬에서 사용하는 모든 객체의 부모

```python
class Person:
    nation = '한국'
    name = str()
    age = int()
    height = float()
    
    def info(self, alias):                            #self는 현재 만들어진 객체의 주소를 받는다
        print('국적 : ', self.nation)
        print('이름 : ', self.name, ', ', alias)
        print('나이 : ', self.age)
        print('키 : ', self.height)
        print('self : ', id(self))

hong = Person()
hong.name = '홍길동'
print('이름 : ', hong.name)
hong.age = 20
hong.height = 170
hong.info('피리 부는 사나이')
print('hong의 주소 : ', id(hong))


p1 = Person()
p1.name = '유비'
print('이름 : ', hong.name)
p1.age = 25
p1.height = 175
p1.info('장사꾼')
print('p1의 주소 : ', id(p1))


Person.nation = '뉴질랜드'
p1.info('산적')

print(dir(Person))
```

```python
class Calculator:
    result = 0
    def __init__(self, *args):
        for i in range(len(args) - 1):
            self.result += args[i]
        print(self.result)

ob1 = Calculator(10, 20, '+')
ob2 = Calculator(10, 20, 30, '+')
```

```python
class Animal:
    def move(self):
        print('움직일 수 있는 동물')
        
        
class Dog(Animal):
    def my(self):
        print('나는 웰시코기에요')
        
class Horse(Animal):
    pass


dog1 = Dog()
dog1.my()
dog1.move()

horse = Horse()
horse.move()


print(isinstance(dog1, Dog))
print(isinstance(dog1, Animal))
print(isinstance(dog1, Horse))
```

```python
class Calculator:
    def __init__(self, first = None, second = None):
        self.first = first
        self.second = second
    
    def setData(self, first = None, second = None):
        self.first = first
        self.second = second
    
    def sum(self):
        result = self.first + self.second
        return result
    
    def mul(self):
        result = self.first * self.second
        return result
    
    def sub(self):
        result = self.first - self.second
        return result
    
    def div(self):
        result = self.first / self.second
        return result
    

class Calculator_v1(Calculator):
    def pow(self):
        result = self.first ** self.second
        return result
    
    def div(self):
        if self.second == 0:
            print('0으로 나눌 수 없습니다.')
            return
        else:
            return self.first / self.second
        
calc1 = Calculator()
calc1.setData(10, 20)
print(calc1.sum())
            
calc2 = Calculator(20, 30)
print(calc2.mul())
calc2.setData(40, 20)
print(calc2.sub())

print('---------------------------------------')

calc3 = Calculator_v1(2, 4)
print(calc3.pow())
calc3.setData(4, 0)
print(calc3.div())
```
